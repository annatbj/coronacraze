```python
import pandas as pd
file=pd.read_excel("coronaglobalstats-kopi.xlsx")
```

```python
#in order to remove commas:
filenod=file.replace(to_replace=",",value="",regex=True)
#in order to make it numeric again we need to go through every file:
filenod["cases"]=pd.to_numeric(filenod["cases"])
filenod["recovered"]=pd.to_numeric(filenod["recovered"],errors="coerce")
filenod["deaths"]=pd.to_numeric(filenod["deaths"], errors="coerce")
```
